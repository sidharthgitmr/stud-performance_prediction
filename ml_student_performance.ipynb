```python
import pandas as pd
```

```python
x = pd.read_csv('Student_Performance.csv')
x.head()
```

```python
c = x.select_dtypes('O')
c.head()
```

```python
x.drop('Extracurricular Activities',inplace=True,axis=1)
```

```python
pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
c.rename({'Extracurricular Activities':'eca'},axis=1,inplace=True)
```

```python
c
```

```python
encoded=le.fit_transform(c)
```

```python
encoded
```

```python
c["eca"]=encoded
```

```python
c
```

```python
fg=pd.concat([x,c],axis=1) 
```

```python
fg
```

```python
x = fg.drop(['Performance Index'],axis=1)
```

```python
x
```

```python
y = fg['Performance Index']
y
```

## scaling using standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xscaled=scaler.fit_transform(x)
```

```python
xscaled
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xscaled,y,test_size=0.2,random_state=32)
```

## model building

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
ypred=model.predict(xtest)
```
